```python
import unittest
from bookcatalog import BookCatalog

class TestExercise(unittest.TestCase):
    
    def test_addBook_succesAddBook(self):
        book = BookCatalog('Title', 'Author', 'Identifier')
        
        book.addBook()
        
        self.assertEqual(BookCatalog.books[-1], ['Title', 'Author', 'Identifier'])
        
    def test_addBook_failedAddBook(self):
        book = BookCatalog('Title', 'Author', 'Identifier')
        book.addBook()
        
        book1 = BookCatalog('Title1', 'Author1', 'Identifier')
        result = book1.addBook()
        
        self.assertEqual(result, "This book already exists")
        
    def test_searchBook_succesSearchBook(self):
        book = BookCatalog('Title', 'Author', 'Identifier')
        book.addBook()
        
        result = book.searchBook('Title')
        
        self.assertEqual(result, ['Title', 'Author', 'Identifier'])

    def test_searchBook_failedSearchBook(self):
        book = BookCatalog('Title', 'Author', 'Identifier')
        book.addBook()
        
        result = book.searchBook('Title1')  
             
        self.assertEqual(result, "This book is not found in the catalog")
        
    def test_removeBook_succesRemoveBook(self):
        book1 = BookCatalog('Title1', 'Author1', 'Identifier1')
        book2 = BookCatalog('Title2', 'Author2', 'Identifier2')

        book1.addBook()
        book2.addBook()
        
        book2.removeBook('Title2')  
        
        self.assertEqual(BookCatalog.books, [['Title1', 'Author1', 'Identifier1']] )
        
    def test_removeBook_failedRemoveBook(self):
        book = BookCatalog('Title', 'Author', 'Identifier')

        book.addBook()
        
        result = book.removeBook('Title2')  
        
        self.assertEqual(result, "This book is not found in the catalog")



        
        

        

        

if __name__ == '__main__':
    unittest.main(argv=['ignored', '-v'], exit=False)
    
```
