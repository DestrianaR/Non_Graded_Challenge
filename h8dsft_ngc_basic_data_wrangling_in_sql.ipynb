```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()
project_id = "first-project-412206"
client = bigquery.Client(project=project_id)
```

```python
query = """
  SELECT table_name
  FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES`
"""

query_job = client.query(query).to_dataframe()

query_job
```

```python
query1 = """
  SELECT column_name, data_type
  FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
  WHERE (table_name = 'bikeshare_stations'
    AND data_type= "STRING");
"""

query_job1 = client.query(query1).to_dataframe()

query_job1
```

```python
query2 = """
  SELECT trip_id, duration_minutes
  FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
  WHERE (trip_id IS NOT NULL
    AND duration_minutes > 1000)
"""

query_job2 = client.query(query2).to_dataframe()

query_job2
```

```python
query3 = """
  SELECT trip_id, subscriber_type, end_station_name
  FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
  WHERE (subscriber_type = "Walk Up"
  AND end_station_name = "Stolen")
"""

query_job3 = client.query(query3).to_dataframe()

query_job3
```
